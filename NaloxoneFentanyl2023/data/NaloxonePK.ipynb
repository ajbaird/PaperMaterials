```python
t = "Time(s)"
y1 = "VenaCava-Naloxone-Concentration(ug/L)"
y2 = "VenaCava-Fentanyl-Concentration(ug/L)"
```

```python
#creating a plotting function so we don't need to call this over and over
def prettyPlot(valData, simultatedData, errData, scalet = True, scaleVal = True, saveName = ""):
    plt.clf()
    fig1 = plt.figure()
    ax = plt.gca()
    plt.rc('font', family='serif')
    plt.rc('xtick', labelsize = 'medium')
    plt.rc('ytick', labelsize = 'medium')
    # set the size 
    fig1 = plt.figure(figsize = (6.472, 4))
    #get the axsis 
    ax = fig1.add_subplot(1,1,1)
    #use a nice stylesheet
    plt.style.use('seaborn-whitegrid')
    #add axis labeling
    ax.set_xlabel('Time (min)')
    ax.set_ylabel('Venous Naloxone Concentration (ug/L)')

    if scalet:
        plt.plot(simultatedData[t]/60, simultatedData[y1])
    else: 
        plt.plot(simultatedData[t], simultatedData[y1])
    if scaleVal:
        plt.errorbar(valData[:,0], valData[:,1] / 1e3, yerr=errData.T, fmt='D', color='violet')
    else: 
        plt.errorbar(valData[:,0], valData[:,1], yerr=errData.T, fmt='D', color='grey')

    if saveName:
        fig1.savefig(saveName + ".png", dpi=500)
```

```python
#creating a plotting function so we don't need to call this over and over
def prettyPlotCompare(valData1,valData2, simulatedData1, simulatedData2, errData1, 
errData2, scalet = True, scaleVal = True, saveName = ""):
    plt.clf()
    fig1 = plt.figure()
    ax = plt.gca()
    plt.rc('font', family='serif')
    plt.rc('xtick', labelsize = 'medium')
    plt.rc('ytick', labelsize = 'medium')
    # set the size 
    fig1 = plt.figure(figsize = (6.472, 4), dpi=500)
    #get the axsis 
    ax = fig1.add_subplot(1,1,1)
    #use a nice stylesheet
    plt.style.use('seaborn-whitegrid')
    #add axis labeling
    ax.set_xlabel('Time (min)')
    ax.set_ylabel('Venous Naloxone Concentration (ug/L)')

    if scalet:
        plt.plot(simulatedData1[t]/60, simulatedData1[y1])
        plt.plot(simulatedData2[t]/60, simulatedData2[y1])
    else: 
        plt.plot(simulatedData1[t], simulatedData1[y1])
        plt.plot(simulatedData2[t], simulatedData2[y1])
    if scaleVal:
        plt.errorbar(valData1[:,0], valData1[:,1] / 1e3, yerr=errData1.T, fmt='D', color='violet')
        plt.errorbar(valData2[:,0], valData2[:,1] / 1e3, yerr=errData2.T, fmt='o', color='grey')
    else: 
        plt.errorbar(valData1[:,0], valData1[:,1], yerr=errData1.T, fmt='D', color='violet')
        plt.errorbar(valData2[:,0], valData2[:,1], yerr=errData2.T, fmt='o', color='grey')

    if saveName:
        fig1.savefig(saveName + ".png", dpi=500)
```
