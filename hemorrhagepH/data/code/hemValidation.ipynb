```python
def moving_average(x, window_size):
    return np.convolve(x, np.ones(window_size) / window_size, mode='valid')

def curvature(x, y):
    dx_dt = np.gradient(x)
    dy_dt = np.gradient(y)
    dx_dt[dx_dt == 0] = 1e-6  # Avoid division by zero
    dy_dt[dy_dt == 0] = 1e-6  # Avoid division by zero
    d2x_dt2 = np.gradient(dx_dt)
    d2y_dt2 = np.gradient(dy_dt)
    denominator = (dx_dt**2 + dy_dt**2)**(3/2)
    curvature = np.abs(d2x_dt2 * dy_dt - dx_dt * d2y_dt2) / denominator
    curvature[np.isnan(curvature)] = 0  # Set curvature to zero where denominator is zero
    return curvature

def highest_curvature_indices(x, y, num_points, tolerance, disregard_start, disregard_end):
    curvatures = curvature(x, y)
    boundary_indices = np.concatenate((np.arange(disregard_start), np.arange(len(x) - disregard_end, len(x)))) # Indices to disregard
    sorted_indices = np.argsort(curvatures)[::-1]  # Sort indices in descending order of curvature
    highest_indices = []
    for index in sorted_indices:
        if index in boundary_indices:
            continue  # Skip boundary indices
        flag = True
        for i in highest_indices:
            if np.abs(x[index] - x[i]) < tolerance and np.abs(y[index] - y[i]) < tolerance:
                flag = False
                break
        if flag:
            highest_indices.append(index)
        if len(highest_indices) >= num_points:
            break
    return highest_indices

def plot_trajectory_with_highest_curvature_points(x, y, num_points, tolerance, disregard_start, disregard_end, window_size):
    # Smooth the trajectory
    x_smoothed = moving_average(x, window_size)
    y_smoothed = moving_average(y, window_size)
    
    plt.figure(figsize=(8, 6))
    plt.plot(x_smoothed, y_smoothed, label='Smoothed Trajectory')
    highest_indices = highest_curvature_indices(x_smoothed, y_smoothed, num_points, tolerance, disregard_start, disregard_end)
    highest_points = [(x_smoothed[i], y_smoothed[i]) for i in highest_indices]
    for point in highest_points:
        plt.scatter(*point, color='red', label='Highest Curvature Point')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Smoothed Trajectory with Highest Curvature Points')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
x =  df.loc['HowToHemValidationpH200.csv', 'BloodVolume(mL)'].to_numpy()
y = df.loc['HowToHemValidationpH200.csv', 'VenaCava-Lactate-Molarity(mmol/mL)'].to_numpy()

num_points = 1  # Set number of highest curvature points to plot
tolerance = 100  # Set tolerance for high curvature points
disregard_start = 1000  # Set number of boundary points to disregard from the start
disregard_end = 500  # Set number of boundary points to disregard from the end
window_size = 5  # Set window size for smoothing
plot_trajectory_with_highest_curvature_points(x, y, num_points, tolerance, disregard_start, disregard_end, window_size)
```
